```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_the_most_common_passwords'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find_all('table')[1]
```

```python
pass_titles = table.find_all('th')
```

```python
password_title_list = [title.text.strip() for title in pass_titles]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = password_title_list)
df
```
